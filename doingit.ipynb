## IMPORTING LIBRARYS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## READING DATASET

```python
df_pokemon=pd.read_csv('data/pokemon.csv')
df_pokemon
```

## MOST COMMON ABILITIES

```python
df_abilities=df_pokemon.copy()
df_abilities['abilities']=df_abilities['abilities'].apply(lambda x: x.replace('[','').replace(']','').replace("'","").split(','))
df_abilities=df_abilities.explode('abilities')
```

```python
df_abilities
```

```python
abilities_count=pd.DataFrame(df_abilities['abilities'].value_counts()).reset_index().rename(columns={"index":"values"}).iloc[0:15]
abilities_count.columns=['abilities','values']
sns.set(rc={"figure.dpi":100})
plt.rcParams['figure.figsize'] = [20, 10]
plot=sns.barplot(data=abilities_count,x='abilities',y='values')
plot.bar_label(plot.containers[0])
plot.set_title('15 most common abilities')
plt.show()
```

## POKEMON BY TYPE

```python
df_types=df_pokemon.copy()
df_types['type2']=df_types['type2'].fillna('')
#df_types['type']=(df_types['type1'] + ',' + df_types['type2']).apply(lambda x: x.strip().split(","))
df_types['type']=(df_types['type1'] + ' ' + df_types['type2']).apply(lambda x: x.strip().split(" ")).apply(lambda x: [x[0],x[1].lstrip()] if len(x) > 1 else x)
df_types=df_types.drop(columns=['type1','type2'])
df_types=df_types.explode('type')
df_types
```

```python

plt.rcParams['figure.figsize'] = [20, 10]
cm = sns.color_palette("rainbow",18) #Splitting the palette in the same amount of numbers bins
plot=sns.histplot(data=df_types['type'])
plot.bar_label(plot.containers[0])
plot.set_title('pokemon amount by type(single or dual)')
for bin_,i in zip(plot.patches,cm):
    bin_.set_facecolor(i)
plt.show()
```

## LEGENDARY BY TYPE

```python
legendary_types_count=pd.DataFrame(df_types[df_types['is_legendary'] == 1]['type'].value_counts()).reset_index()
legendary_types_count.columns=['type','count']
legendary_types_count
```

```python
plt.rcParams['figure.figsize'] = [20, 10]
cm = sns.color_palette("rainbow",18) #Splitting the palette in the same amount of numbers bins
plot=sns.barplot(data=legendary_types_count,y='count',x='type')
plot.bar_label(plot.containers[0])
plot.set_title('legendary pokemon by type')
for bin_,i in zip(plot.patches,cm):
    bin_.set_facecolor(i)
plt.show()
```

## dividing pokemon by base stat total

main five divisions will be:
very low(<250)
low(250 -400)
medium(401-550)
high(551-679)
very high(> 679)

```python

very_low=df_pokemon[(df_pokemon['base_total'] >= 180) & (df_pokemon['base_total'] < 250)]
low=df_pokemon[(df_pokemon['base_total'] >= 250) & (df_pokemon['base_total'] < 401)]
medium=df_pokemon[(df_pokemon['base_total'] >= 401) & (df_pokemon['base_total'] < 551)]
high=df_pokemon[(df_pokemon['base_total'] >= 551) & (df_pokemon['base_total'] < 681)]
very_high=df_pokemon[df_pokemon['base_total'] > 679]
byBaseStat=pd.DataFrame(
    index=['very low','low','medium','high','very high'],
    columns=['count','amt_legendary'],
    data=zip(
            [very_low['abilities'].count(),low['abilities'].count(),medium['abilities'].count(),high['abilities'].count(),very_high['abilities'].count()],
            [very_low[very_low['is_legendary'] == 1]['abilities'].count(),low[low['is_legendary'] == 1]['abilities'].count(),medium[medium['is_legendary'] == 1]['abilities'].count(),high[high['is_legendary'] == 1]['abilities'].count(),very_high[very_high['is_legendary'] == 1]['abilities'].count()])
)
byBaseStat=byBaseStat.reset_index()
byBaseStat.columns=['category','count','amt_legendary']
byBaseStat
```

```python
plt.rcParams['figure.figsize'] = [20, 10] #Splitting the palette in the same amount of numbers bins
sns.barplot(data=byBaseStat,x='category',y='count',hue='amt_legendary',palette='Spectral',dodge=False)
plot.set_title('amount of pokemon by total base stat')
plt.show()
```
